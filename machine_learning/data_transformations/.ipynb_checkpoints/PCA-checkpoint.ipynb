**Метод главных компонент**

Все особенности и тонкоссти исползования метода главных компонент.

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
```

# Первая главная компонента

## *Определение*

Пусть имеюстся $n$ наблюдений за $m$ переменными - $x_{ij}, i\in \overline{1,n}, j \in \overline{1,m}$. Тогда первой главной компонентой назвается функция:

$$Z(x_1, x_2, ... , x_m) = \sum_{j=1}^m\phi_{j}(x_j - \bar{x}_j);$$

Где:
- $\bar{x}_j$ - среднее выборочное $j$-й переменной $\forall j$.

При том $\phi_j$ подбираются так, чтобы:

$$(\phi_1, \phi_2, ... \phi_m) = argmin_{\varphi_1, \varphi_2, ..., \varphi_{m}}\left\{Var\left[\sum_{j=1}^m\varphi_{j}(x_j - \bar{x}_j)\right]\right\};$$

$$\sum_{j=1}^m \varphi_j^2 = 1.$$

## *Пример*

Допустим имеются $n$ наблюдений за 2-мя переменными.

```python
n = 500

plt.figure(figsize = [14, 7])

np.random.seed(10)
x1 = np.random.rand(n)
x2 = x1 + 5 + np.random.normal(0, 0.4, n)

pca = PCA(n_components=1).fit(
    np.concatenate(
        [x1[:, np.newaxis], x2[:, np.newaxis]], axis = 1
    )
)
x1_arange = np.arange(0,1.001, 0.1)
pca1_coefs = pca.components_[0]
pca1_arange = (pca1_coefs[0]*x1_arange)/pca1_coefs[1]

ans = plt.scatter(x1, x2)
plt.plot(x1_arange, pca1_arange, color = "green")
plt.xlabel("$x_1$", fontsize = 14)
plt.ylabel("$x_2$", fontsize = 14)
```

Коэффиценты первой главной компонент составят:

```python
pca = PCA(n_components=1).fit(
    np.concatenate(
        [x1[:, np.newaxis], x2[:, np.newaxis]], axis = 1
    )
)
pca.components_
```
