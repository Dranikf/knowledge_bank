**Компромисс дисперсии и смещения**

В этом notebook я надеюсь подробно разобрать вопрос компромиса между смещением (bias) и дисперсией (variance) в машинном обучении.

Вдохновлено соотсветсвующим разделом в <a href="https://www.statlearning.com/">ISLR</a>.

# Введение

Утверждается что средне квадратическую ошибку модели (Mean Square Error, MSE) можно разложить на три составляющие:
- Дисперсию;
- Квардрат симещения;
- Неустранимую ошибку.

Или записывая серез формулу:

$$\mathbb{E}(y_0 - \hat{f}(x_0))^2 = Var(\hat{f}(x_0)) + [Bias(\hat{f}(x_0))]^2 + Var(\varepsilon).$$

Где:
- $\hat{f}(x)$ - некоторая модель описывающая данные;
- $x_0$ - некоторое произвольное контрольное значение предикторов;
- $y_0$ - некоторое произвольное контрольное значение оклика;
- $\varepsilon$ - неустранимая ошибка модели; в нее вкладывается та информация о описываемом процессе, которой нет в полученной выборке;
- $Var(\hat{f}(x_0))$ - дисперсия полученной модели; эта величина должна быть тем большем, чем сильнее предстказания модели будут меняться от измениния выбоки;
- $Bias(\hat{f}(x_0))$ - смещение меодли эта величина тем меньше, чем точнее модель подогнанна к описываемому просцессу;
- $Var(\varepsilon)$ - дисперсия неустранимой ошибки; чем полнее и точнее наши данные, тем меньше эта величина;
- $\mathbb{E}(y_0 - \hat{f}(x_0))^2$ - математическое ожидание ошибки полученной модели.

# Каждый компонент ошибки на примере

### Неустранимая ошибка

Допустим, есть некоторый процесс, который мы хотим описать регрессионной моделью. На практике, процесс всегда не известен (в противном случае мы бы просто не нуждались в модели для его описания). На процесс, всегда оказывает влияние очень большое число факторов, но мы смогли собрать информацию только об некоторых из них, обычно тех, которые, из соображений логики, влияют на него сильнее всего и, при том, их относительно легко собрать. Влияние остальных факторов а также ошибки, допущенные при сборе инфомации, формируют неустранимую ошибку ($\varepsilon$).

Допустим, мы хотим предсказывать некоторый $y \in \mathbb{R}$. Но собрать информацию нам удалось только об некоторой перменной $x \in \mathbb{R}$. Если бы $y$ зависил только от $x$, то возможно было бы подобрать такую $f$, что:

$$y = f(x).$$

То чаще всего на $y$ влияют и другие переменные, а о них в результате несовершенства нашей информации мы ничего не знаем. Так и получается, что лучшее на что мы можем рассчитывать это:

$$y=f(x)+\varepsilon.$$

Т.е. в $\varepsilon$ вкладывается та информация которую мы никак не можем восстановитьт используя только $x$.
